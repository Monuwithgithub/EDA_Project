import necessary libraries 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Load the Csv File

```python
df=pd.read_csv("Books_Data_Clean.csv")
```

```python
df.head()
```

Data Cleaning 


```python
df.describe()
```

```python
df = df[df["Publishing Year"]>1900]
```

```python
df.isna().sum()
```

```python
df.dropna(subset= "Book Name", inplace= True)
```

```python
df.duplicated().sum()
```

```python
df.nunique()
```

Exploratory Analysis

```python
plt.hist(df['Publishing Year'])
plt.xlabel('Publishing Year')
plt.ylabel('Frequency')
plt.title("Distribution of Publication Year")
plt.show()
```

```python
df["genre"].value_counts().plot(kind="bar")
plt.xlabel("Genre")
plt.ylabel("Number of Books")
plt.title("Nember of Books in Each Genre")
plt.show()
```

```python
df.groupby("Author")["Book_average_rating"].mean().sort_values(ascending= False)
```

```python
sns.boxplot(x="genre" , y="Book_ratings_count", data=df)
plt.xlabel("Genre")
plt.ylabel("Book Ratings Count")
plt.title("Box plot of book ratings count for each genre" )
plt.show()
```

```python
plt.scatter(df["sale price"], df["units sold"])
plt.xlabel("Sales Price")
plt.ylabel("Units Sold")
plt.title("Scatter plot of sale price vs Units Sold")
plt.show()
```

```python
language_counts= df["language_code"].value_counts()
```

```python
plt.pie(language_counts, labels=language_counts.index, startangle=90)
plt.title("Language Distribution of Books")
plt.show()
```

```python
total_grass_sales_by_author = df.groupby("Author") ["gross sales"].sum()
```

```python
total_grass_sales_by_author.sort_values(ascending= False).head(10)
```

```python
total_grass_sales_by_author.sort_values(ascending= False).head(10).plot(kind="bar")
plt.xlabel("Author")
plt.ylabel("Total gross sales")
plt.title("Total gross sales for each author")
plt.show()
```

```python
sns.boxplot(x= "Author_Rating", y= "units sold", data=df)
plt.xlabel("Author Rating")
plt.ylabel("Units Sold")
plt.title("Box plot of units sold for each author rating")
plt.show()
```

```python
df.groupby("Publishing Year")["units sold"].sum().plot(kind="line", marker="o")
plt.xlabel("Publishing year")
plt.ylabel("Total units sold")
plt.title("Total unit sold over the year")
plt.show()
```
